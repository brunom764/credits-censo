```python
def retornaRGB(imagem,y,x):
  (b, g, r) = imagem[y, x] #veja que a ordem BGR e não RGB
  print(f'O pixel ({y}, {x}) tem as seguintes cores:')
  print(f'Vermelho:{r}, Verde:{g}, Azul:{b}')

retornaRGB(imagem2,10,12)
```

```python
def inverteRGB(imagem):
  for y in range(0, imagem.shape[0]):
    for x in range(0, imagem.shape[1]):
      (b, g, r) = imagem[y, x]
      imagem[y, x] = (r, g, b)
      
  cv2_imshow(imagem)

inverteRGB(imagem2)
```

```python
def gray(imagem):
  imagem_cinza = cv2.cvtColor(imagem2, cv2.COLOR_BGR2GRAY)
  cv2_imshow(imagem_cinza)

gray(imagem2)
```

```python
import numpy as np
def converteCinza(img):
  (canalAzul, canalVerde, canalVermelho) = cv2.split(img)
  cv2_imshow(canalVermelho/3)
  cv2_imshow(1.75*canalVerde/3)
  cv2_imshow(1.4*canalAzul/3)
  cv2.waitKey(0)


converteCinza(imagem2)
```
